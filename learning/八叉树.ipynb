```python
# 八叉树
import open3d as o3d
from sympy.external.gmpy import invert

pcd = o3d.io.read_point_cloud("1.ply")
# 建立八叉树
octree = o3d.geometry.Octree(max_depth=10) #设置最大深度
octree.convert_from_point_cloud(pcd,size_expand=0.1) #size_expand叶子节点大小
o3d.visualization.draw_geometries([octree],window_name="八叉树",width=800,height=600)
```

```python
# 点云相交 求异
import open3d as o3d
import numpy as np
# 读取原始点云和求异点云
pc1 = o3d.io.read_point_cloud("2.ply", remove_nan_points=True, remove_infinite_points=True)
pc2 = o3d.io.read_point_cloud("5.ply", remove_nan_points=True, remove_infinite_points=True)

# 建立原始点云数据的kd树
k_dTree = o3d.geometry.KDTreeFlann(pc1)
ptsIdx = [] #两个点云中相同部分的索引
k = 1 #k是knn搜索的参数，也就是说搜索另一片点云中距离它的最近点
dist_max = 0.1 #距离阈值设置为0.1
points = np.array(pc2.points)
pointNum = points.shape[0]
# 遍历点云
for i in range(0, pointNum):
    # k 返回点的个数
    # idx 返回点的索引
    # dist 返回点的距离
    [k,idx,dist] = k_dTree.search_knn_vector_3d(pc2.points[i],k) #通过kd树进行搜索最近点
    if dist[0] < dist_max: #如果另一片点云中能找到距离小于给定阈值的点，则判定为相同部分
        ptsIdx.append(i)

# 最后将点云中相同的部分和不同的部分分别取出显示
same_part = pc2.select_by_index(ptsIdx)
diff_part = pc2.select_by_index(ptsIdx,invert=True)
same_part.paint_uniform_color([0,0,1])
diff_part.paint_uniform_color([0,1,0])
o3d.visualization.draw_geometries([same_part,diff_part])
```
