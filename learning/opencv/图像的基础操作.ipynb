# 1. 图片的IO操作

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 读取图片
horse = cv.imread('img/horse.jpg')

# 保存图片
cv.imwrite('img/horseCopy.png',horse)
```

```python
# opencv展示图片
cv.imshow('horse picture', horse)
cv.waitKey(0) # 一直等待
cv.destroyAllWindows() # 当有键盘输入时销毁图片
```

```python
# matplotlib展示图片
horse = plt.imshow(horse[:,:,::-1])
plt.show()
```

# 2. 绘制图形/文字

```python
img = np.zeros((512,512,3), np.uint8)

# 绘制图形
cv.line(img,(0,0),(512,512),(255,0,0),thickness=5)
cv.circle(img,(100,100),20,(0,255,0),thickness=5)
cv.rectangle(img,(400,400),(512,512),(0,0,255),thickness=5)
# 添加文字
cv.putText(img,'Hello',(200,100),cv.FONT_HERSHEY_SIMPLEX,2,(0,0,255),thickness=2)

# 显示结果
plt.imshow(img[:,:,::-1])
```

# 2.numpy读取图片存储格式

```python
# 创建一个2×2的简单图像
img = np.zeros((2, 2, 3), dtype=int)

# 给每个通道赋值
img[:,:,0] = [[1, 20], [3, 4]]    # 红色通道
img[:,:,1] = [[10, 20], [30, 40]] # 绿色通道
img[:,:,2] = [[100, 200], [300, 400]] # 蓝色通道

print("完整3D数组:")
print(img)
plt.imshow(img)
"""
[[[  1  10 100]   # 位置(0,0): R=1, G=10, B=100
  [  2  20 200]]  # 位置(0,1): R=2, G=20, B=200

 [[  3  30 300]   # 位置(1,0): R=3, G=30, B=300
  [  4  40 400]]] # 位置(1,1): R=4, G=40, B=400
"""
```

```python
print("\n红色通道（第0层）:")
print(img[:,:,0])  # 只看R通道
"""
[[1 2]
 [3 4]]
"""
```

```python
print("\n绿色通道（第1层）:")
print(img[:,:,1])  # 只看G通道
"""
[[10 20]
 [30 40]]
"""
```

```python

print("\n蓝色通道（第2层）:")
print(img[:,:,2])  # 只看B通道
"""
[[100 200]
 [300 400]]
"""
```

```python
# 现在反转通道顺序
img_reversed = img[:,:,::-1]
print("\n反转后的完整数组:")
print(img_reversed)
"""
[[[100  10   1]   # 原来(0,0)的[1,10,100] → [100,10,1]
  [200  20   2]]  # 原来(0,1)的[2,20,200] → [200,20,2]

 [[300  30   3]   # 原来(1,0)的[3,30,300] → [300,30,3]
  [400  40   4]]] # 原来(1,1)的[4,40,400] → [400,40,4]
"""
```
