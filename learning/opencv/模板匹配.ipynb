# 1.模板匹配
实现原理： 在原图中找到与模板相匹配的区域

```python
from tempfile import template

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# 加载原图
img = cv.imread("img/wulin.jpeg")        # BGR
img = cv.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# 加载模板
template = cv.imread('img/bai.jpeg')
template = cv.cvtColor(template, cv.COLOR_BGR2RGB)
plt.imshow(template)
```

```python
# 进行模板匹配
res = cv.matchTemplate(img,template,cv.TM_CCORR)
plt.imshow(res,cmap='gray')
```

```python
# 解析模板匹配结果
min_val, max_val,min_loc,max_loc = cv.minMaxLoc(res)
top_left = max_loc # 拿到原图中与模板最匹配的位置
# 确定右下角位置
h,w = template.shape[:2] # 拿到模板高宽 h w
bottom_right = (top_left[0] + w, top_left[1] + h)
# 在原图上绘制模板的矩形框
cv.rectangle(img,top_left,bottom_right,(0,255,0),2)
plt.imshow(img)
```
